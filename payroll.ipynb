```python
worked_hours = float(input("Put how many hours you worked in the month: "))
hour_value = float(input("Put what your hour is worth: "))
gross_salary = float(worked_hours * hour_value)
print (">>>>>>>>>> ( = ) SALÁRIO BRUTO: {:.2f}".format(gross_salary))
union_dues = gross_salary * 0.05
print (">>>>>>>>>> ( - ) CONTRIBUIÇÃO SINDICAL: {:.2f} ".format(union_dues))
fgts = gross_salary * 0.08
print (">>>>>>>>>> ( - ) FGTS: {:.2f}".format(fgts))

if gross_salary <= 1212.00:
  aliquot = 0.075
  deductible_installment = 0.00
  inss_discount = (gross_salary * aliquot) - deductible_installment
  print (">>>>>>>>>> ( - ) I.N.S.S.: {:.2f}".format(inss_discount))

elif gross_salary >= 1212.01 and gross_salary <= 2427.35:
  aliquot = 0.09
  deductible_installment = 18.185
  inss_discount = (gross_salary * aliquot) - deductible_installment
  print (">>>>>>>>>> ( - ) I.N.S.S.: {:.2f}".format(inss_discount))

elif gross_salary >= 2427.36 and gross_salary <=3641.03:
  aliquot = 0.12
  deductible_installment = 91.00
  inss_discount = (gross_salary * aliquot) - deductible_installment
  print (">>>>>>>>>> ( - ) I.N.S.S.: {:.2f}".format(inss_discount))

elif gross_salary >=3641.04 and gross_salary <= 7087.22:
  aliquot = 0.14
  deductible_installment = 163.83
  inss_discount = (gross_salary * aliquot) - deductible_installment
  print (">>>>>>>>>> ( - ) I.N.S.S.: {:.2f}".format(inss_discount))

elif gross_salary >= 7087.23:
  inss_discount = 828.38
  print (">>>>>>>>>> ( - ) I.N.S.S.: {:.2f}".format(inss_discount))

else:
  print ("ERROR ")

salary_m_inss = (gross_salary - inss_discount)

if salary_m_inss >= 0.00 and salary_m_inss <= 1903.98:
  irrf_discount = 0.00
  print (">>>>>>>>>> ( - ) I.R.R.F.: {:.2f}".format(irrf_discount))
  net_salary = salary_m_inss - union_dues - irrf_discount
  print ("\n>>>>>>>>>> ( = ) NET SALARY: {:.2f}".format(net_salary))

elif salary_m_inss >=1903.99 and salary_m_inss <= 2826.65:
  aliquot = 0.075
  deductible_installment = 142.80
  irrf_discount = ((gross_salary - inss_discount) * aliquot) - deductible_installment
  print (">>>>>>>>>> ( - ) I.R.R.F.: {:.2f}".format(irrf_discount))
  net_salary = salary_m_inss - union_dues - irrf_discount
  print ("\n>>>>>>>>>> ( = ) NET SALARY: {:.2f}".format(net_salary))

elif salary_m_inss >= 2826.66 and salary_m_inss <= 3751.05:
  aliquot = 0.15
  deductible_installment = 354.80
  irrf_discount = ((gross_salary - inss_discount) * aliquot) - deductible_installment
  print (">>>>>>>>>> ( - ) I.R.R.F.: {:.2f}".format(irrf_discount))
  net_salary = salary_m_inss - union_dues - irrf_discount
  print ("\n>>>>>>>>>> ( = ) NET SALARY: {:.2f}".format(net_salary))

elif salary_m_inss >= 3751.06 and gross_salary <= 4664.68:
  aliquot = 0.2250
  deductible_installment = 636.13
  irrf_discount = ((gross_salary - inss_discount) * aliquot) - deductible_installment
  print (">>>>>>>>>> ( - ) I.R.R.F.: {:.2f}".format(irrf_discount))
  net_salary = salary_m_inss - union_dues - irrf_discount
  print ("\n>>>>>>>>>> ( = ) NET SALARY: {:.2f}".format(net_salary))

else:
  salary_m_inss >= 4664.69
  aliquot = 0.2750
  deductible_installment = 869.36
  irrf_discount = ((gross_salary - inss_discount) * aliquot) - deductible_installment
  print (">>>>>>>>>> ( - ) I.R.R.F.: {:.2f}".format(irrf_discount))
  net_salary = salary_m_inss - union_dues - irrf_discount
  print ("\n>>>>>>>>>> ( = ) NET SALARY: {:.2f}".format(net_salary))
```
